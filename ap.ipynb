```python
import numpy as np 

data=np.linspace(1,10,9,endpoint=False)
data=np.reshape(data,(3,3))

imposmoves=[]
finalset=set()
l=[]
pos_patterns=[]
startdigit=2
lastdigit=2

#finding possible moves
for r in range(0,3):
    for c in range(0,3):
        digit_impos_moves=set()
        for co in range(0,3):
            digit_impos_moves.add(data[r,co])
        for ro in range(0,3):
            digit_impos_moves.add(data[ro,c])
        imposmoves.append(digit_impos_moves)
#finding immpossible moves
def findpos_moves(imposmoves):
    s=set(np.arange(1,10))
    pos_moves=[]
    for d in imposmoves:
        pos_moves.append(list(s.difference(d)))
    return pos_moves

pos_moves=findpos_moves(imposmoves)

print(pos_moves)
```

```python
import matplotlib.pyplot as plt
figure=plt.figure(figsize=(8,8))
plt.plot([1,1,1,3,3,3,5,5,5],[1,3,5,1,3,5,1,3,5],"ro")
plt.axis([-1,7,-1,7])

plot_xylist=[ [1,5],[3,5],[5,5],[1,3],[3,3],[5,3],[1,1],[3,1],[5,1] ]    
```

```python
def get_xyvalue(digit):
    return plot_xylist[digit-1]
def plot_2points(last,current):
    lin=plt.plot([last[0],current[0]],[last[1],current[1]],marker="o")
    return lin

#Recursive depth first (Android Pattern Design)
def patterndesign(digit):
    if digit not in finalset:
        finalset.add(digit)
        l.append(digit)
        global lastdigit
        last=get_xyvalue(lastdigit)
        current=get_xyvalue(digit)
        
        # plot last and current
        lin=plot_2points(last,current)
        
        #change last digit as current
        lastdigit=digit
        
        for d in pos_moves[digit-1]:
            patterndesign(d)
        if len(finalset) is not 9:
            finalset.remove(digit)
            print("hi")
            print(lin)
            l.remove(digit)
            lin.pop().remove()
        else:
            return
    else:
        return

patterndesign(startdigit)
print(l)
plt.show()
```
